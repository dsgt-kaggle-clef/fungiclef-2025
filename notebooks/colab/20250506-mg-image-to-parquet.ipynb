### image-to-parquet workflow

```python
from pathlib import Path
from fungiclef.preprocessing.serialize import add_serialized_images_to_csv

data_path = "/content/drive/MyDrive/fungiclef-2025/data"
csv_path = f"{data_path}/metadata/FungiTastic-FewShot/FungiTastic-FewShot-Train.csv"
print(csv_path)

image_dir = f"{data_path}/images/FungiTastic-FewShot/train/fullsize"
print(image_dir)

output_path = f"{data_path}/processed/train_serialized.parquet"
print(output_path)

image_col = "filename"

# create output directory if it doesn't exist
Path(output_path).parent.mkdir(parents=True, exist_ok=True)
```

```python
# create parquet files
add_serialized_images_to_csv(
    csv_path=csv_path,
    image_dir=image_dir,
    output_parquet=output_path,
    image_col=image_col,
)
```
